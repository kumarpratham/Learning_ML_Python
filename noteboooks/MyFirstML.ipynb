```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# Create a simple dataset
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])

print(X)
print(y)
```

```python
# Plot the dataset
plt.plot(X, y, marker='o')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Linear Relationship')
plt.show() 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
predictions = model.predict(np.array([[6], [7], [8], [9], [10]]))
print(predictions)
```

```python
print(model.predict([[20]]))
```

```python
plt.scatter(X, y)

plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Training Data")

plt.show()
```

```python
y_pred = model.predict(X)
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Predicted Line')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Linear Regression Model Fit")
plt.legend()
plt.show()
```

```python
# New dataset
X_new = np.array([[1], [2], [3], [4], [5]])
y_new = np.array([3, 5, 7, 9, 11])

new_model = LinearRegression()
new_model.fit(X_new,y_new)
```

```python
y_new_pred = new_model.predict(X_new)
plt.plot(X_new,y_new_pred, color='red', label='New Predicted Line')
plt.scatter(X_new,y_new, label='Actual Data')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("New Linear Regression Model Fit")
plt.legend()
plt.show()
```

```python
import pandas as pd
```

```python
data = {
    "size": [1000, 1500, 1800, 2400, 3000],
    "rooms": [2, 3, 3, 4, 4],
    "price": [200000, 300000, 360000, 480000, 600000]
}
df = pd.DataFrame(data)
print(df)
```

```python
XX = df[["size", "rooms"]]
yy = df["price"]

print(XX)
print(yy)
```

```python
home_model = LinearRegression()
home_model.fit(XX, yy)
```

```python
print(home_model.predict([[2000, 3]]))
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df["size"], df["rooms"], df["price"])

ax.set_xlabel("House Size")
ax.set_ylabel("Rooms")
ax.set_zlabel("Price")

plt.title("House Price Dataset")

plt.show()
```

```python
size_range = np.linspace(df["size"].min(), df["size"].max(), 10)
rooms_range = np.linspace(df["rooms"].min(), df["rooms"].max(), 10)

size_grid, rooms_grid = np.meshgrid(size_range, rooms_range)

predicted_prices = home_model.predict(
    np.c_[size_grid.ravel(), rooms_grid.ravel()]
).reshape(size_grid.shape)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df["size"], df["rooms"], df["price"], color="blue")
ax.plot_surface(size_grid, rooms_grid, predicted_prices, alpha=0.5)

ax.set_xlabel("Size")
ax.set_ylabel("Rooms")
ax.set_zlabel("Price")

plt.title("Regression Plane")

plt.show()
```
